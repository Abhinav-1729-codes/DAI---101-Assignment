```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

df = pd.read_csv("car_price_dataset.csv")

# Basic info
print("\n Dataset Overview ")
print(df.info())
print("\n Summary Statistics ")
print(df.describe(include='all'))

# Thsi is Univariate Analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

for col in numerical_cols:
    print(f"\n Statistics for {col} ")
    print(df[col].describe())

    # This will be Histogram plot
    plt.figure(figsize=(7, 5))
    sns.histplot(df[col], kde=True, bins=30, color="blue")
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.show()

    # Box Plot
    plt.figure(figsize=(7, 5))
    sns.boxplot(y=df[col], color="green")
    plt.title(f'Box Plot of {col}', fontsize=14)
    plt.ylabel(col, fontsize=12)
    plt.show()

for col in categorical_cols:
    print(f"\n Value Counts for {col} ")
    print(df[col].value_counts())

    plt.figure(figsize=(8, 4))
    sns.countplot(y=df[col], order=df[col].value_counts().index, hue=None, legend=False)
    plt.title(f'Count of {col}', fontsize=14)
    plt.xlabel("Count", fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.show()

# This is Multivariate Analysis
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title("Correlation Matrix", fontsize=14)
plt.show()

if len(numerical_cols) > 1:
    sns.pairplot(df[numerical_cols], diag_kind='kde', corner=True, plot_kws={'alpha': 0.7})
    plt.show()

if len(categorical_cols) > 0 and len(numerical_cols) > 0:
    for cat_col in categorical_cols:
        for num_col in numerical_cols:
            plt.figure(figsize=(8, 4))
            sns.boxplot(y=df[num_col], palette="Set2",legend=False,hue=df[cat_col])
            plt.xticks(rotation=45)
            plt.title(f'{num_col} Distribution Across {cat_col}', fontsize=14)
            plt.xlabel(cat_col, fontsize=12)
            plt.ylabel(num_col, fontsize=12)
            plt.show()

print("\n   Analysis Completed ")
print("\n   Thank You Mam ")
```
